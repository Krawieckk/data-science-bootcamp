```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

Wykres liniowy

```python
_ = plt.plot([0, 1], [0, 3])
```

```python
plt.plot([0, 1], [0, 3], label='line', color='green')
plt.xlabel('os x')
plt.ylabel('os y')
plt.title('Wykres')
_ = plt.legend()
```

```python
plt.plot([0, 1, 2], [0, 3, 2], label='line', color='green')
plt.xlabel('os x')
plt.ylabel('os y')
plt.title('Wykres')
_ = plt.legend()
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
_ = plt.plot(x, x_cos)
```

Tworzenie wielu wykresów

```python
plt.subplot(121)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(122)
plt.plot([0, 1, 2], [3, 1, 4], 'r--')
```

```python
fig = plt.figure(figsize=(12, 8))

plt.subplot(131)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(132)
plt.plot([0, 1, 2], [3, 1, 4], 'r--')
plt.subplot(133)
plt.plot([0, 1, 2], [3, 1, 4], 'ro')
```
